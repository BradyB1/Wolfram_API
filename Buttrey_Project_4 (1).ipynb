```python
#functioning chatbot using Conversational API. 
print("Welcome to chatbot")
#while loop for continous questions
while True:
    #ask if they want to ask a question
    would_you = input("Would you like to ask a question (y or n)? ")
    #check if yes and run
    if would_you.lower() == 'y':
        # Construct the query URL with the units parameter
        question = input("Please enter a question: ")
        query_url = f"http://api.wolframalpha.com/v1/conversation.jsp?" \
                    f"appid={appid}" \
                    f"&i={question}"

        # Send the API request and retrieve the response
        response = requests.get(query_url).json()

        # Extract the answer, conversation ID, and host from the response
        answer = response["result"]
        conversation_id = response["conversationID"]
        host = response["host"]

        # Print the question and its answer
        print(f"{question}: '{answer}'")
        
        #ask if they have follow up questions
        follow_up = input("Would you like to ask a follow-up question? (enter y or n)")
        #if not, break out
        if follow_up.lower() == 'n':
            print("Goodbye!")
            break
        #nested while loop for follow ups
        while True:
            #get the follow-up question
            followup_question = input("Enter Follow-up question: ")

            # Construct the query URL for the follow-up question
            query_url = f"http://{host}/api/v1/conversation.jsp?" \
                        f"appid={appid}" \
                        f"&conversationID={conversation_id}" \
                        f"&i={followup_question}"

            # Send the API request for the follow-up question and retrieve the response
            response = requests.get(query_url).json()

            # Check if "result" key exists in the response
            if "result" in response:
                answer = response["result"]
            else:
                answer = "I'm sorry, I couldn't find the answer to that question."

            # Print the follow-up question and its answer
            print(f"{followup_question}: '{answer}'")
            
            #ask if there are more follow up questions
            follow_up_again = input("Do you have another follow-up question? (y or n)? ")
            #if not, break, otherwise keep running
            if follow_up_again.lower() == "n" or follow_up_again.lower() != 'y':
                break
            
    #close if they have no questions
    elif would_you.lower() == 'n':
        print("Goodbye!")
        break
    #make sure its a valid input
    else:
        raise ValueError("Please enter 'y' for yes or 'n' for no.")
```

```python
#generated information on all of these using full results

inputs = ["capital of France", "population of Germany", "largest city in Japan"]

for input_query in inputs:
    # Encode the input query as a URL parameter
    encoded_query = urllib.parse.quote_plus(input_query)

    # Construct the query URL with the necessary parameters
    query_url = f"http://api.wolframalpha.com/v2/query?" \
                f"appid={appid}" \
                f"&input={encoded_query}" \
                f"&format=plaintext" \
                f"&output=json"

    # Send the API request and retrieve the response
    response = requests.get(query_url).json()

    # Extract relevant information from the response
    pods = response["queryresult"]["pods"]
    print(f"Input: {input_query}\n")

    for pod in pods:
        pod_title = pod["title"]
        subpods = pod["subpods"]

        print(f"Pod Title: {pod_title}")
        for subpod in subpods:
            plaintext = subpod["plaintext"]
            print(f"Plaintext: {plaintext}\n")
```

```python
#generated information on topic using full results
topic = input("Enter a topic: ")

# Encode the topic as a URL parameter
encoded_topic = urllib.parse.quote_plus(topic)

# Construct the query URL with the necessary parameters
query_url = f"http://api.wolframalpha.com/v2/query?" \
            f"appid={appid}" \
            f"&input={encoded_topic}" \
            f"&format=plaintext" \
            f"&output=json"

# Send the API request and retrieve the response
response = requests.get(query_url).json()

# Extract relevant information from the response
pods = response["queryresult"]["pods"]

print(f"Topic: {topic}\n")

for pod in pods:
    pod_title = pod["title"]
    subpods = pod["subpods"]

    print(f"Pod Title: {pod_title}")
    for subpod in subpods:
        plaintext = subpod["plaintext"]
        print(f"Plaintext: {plaintext}\n")
```
